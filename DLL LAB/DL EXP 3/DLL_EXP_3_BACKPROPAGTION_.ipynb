```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Set seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Adjusted dataset: linearly related input and target
x_train = np.array([
    [0.1, 0.2],
    [0.3, 0.7],
    [0.4, 0.5],
    [0.9, 0.6]
])

# Adjusted to better match linear patterns
y_train = np.array([
    0.1 * 0.2 + 0.1 + 0.2,    # 0.32
    0.3 * 0.7 + 0.3 + 0.7,    # 1.21
    0.4 * 0.5 + 0.4 + 0.5,    # 1.10
    0.9 * 0.6 + 0.9 + 0.6     # 2.04
])

# Define a simple model (smaller, avoids overfitting)
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(4, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Function to print weights and biases
def print_weights_and_biases(model, note=""):
    print(f"\nWeights and biases {note}:")
    for i, layer in enumerate(model.layers):
        weights, biases = layer.get_weights()
        print(f"Layer {i+1} weights:\n{weights}")
        print(f"Layer {i+1} biases:\n{biases}")

print_weights_and_biases(model, "before training")

# Train the model (more epochs for better learning)
model.fit(x_train, y_train, epochs=500, verbose=0)

print_weights_and_biases(model, "after training")

# Make predictions
predictions = model.predict(x_train)

# Print predicted vs target values
print("\nPredictions vs Targets:")
for i in range(len(x_train)):
    print(f"Input: {x_train[i]}, Target: {y_train[i]:.2f}, Predicted: {predictions[i][0]:.2f}")
```
